```python
import numpy as np

def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # 지구 반지름 (km 단위)
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1

    a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2
    c = 2 * np.arcsin(np.sqrt(a))
    return R * c
```

```python
# 사건별로 경찰서와의 최소 거리 계산
def min_distance_to_police(lat, lon):
    return min([
        haversine(lat, lon, police_lat, police_lon)
        for police_lat, police_lon in zip(police_df['LATITUDE'], police_df['LONGITUDE'])
    ])

# 적용
crime_df['min_distance_to_police'] = crime_df.apply(
    lambda row: min_distance_to_police(row['Latitude'], row['Longitude']), axis=1
)
```
